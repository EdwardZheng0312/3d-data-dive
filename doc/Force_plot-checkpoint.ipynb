```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Force0010.csv', header=None)
```

```python
data.head()
```

```python
data.shape
```

```python
y1 = data[0]
#y1
```

```python
y2 = data[1]
```

```python
y3 = data[2]
```

```python
y4 = data[3]
```

```python
x = np.linspace(0, len(y1)-1, len(y1) )
x
```

```python
fig = plt.figure(figsize=(36,9))
ax = fig.add_subplot(144)
ax.scatter(x, y4)
plt.title('phase shift')

ax = fig.add_subplot(143)
ax.scatter(x, y3)
plt.ylim(y3.min(), y3.max())
plt.title('amplitude')

ax = fig.add_subplot(142)
ax.scatter(x, y2)
plt.ylim(2.7e-9,3.84e-9)
plt.title('deflection')

ax = fig.add_subplot(141)
ax.scatter(x, y1)
plt.ylim(y1.min(), y1.max())
plt.title('tip_motion')



plt.show()
```
